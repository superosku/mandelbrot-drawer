```python
def get_img(pos, res=size):
    xpos, ypos, zoom = pos
    img = np.zeros((res, res), dtype='float')
    #print(z.shape)
    
    #xsize = x1 - x0
    #ysize = y1 - y0
    
    
    c = np.array([
        [
            (float(zoom) * (float(x) / res - 0.5) - float(xpos)) + 
            (float(zoom) * (float(y) / res - 0.5) - float(ypos)) * 1j
            #(x / size * xsize + x0) + (y / size * ysize + y0) * 1j
            for y in range(res)
        ]
        for x in range(res)
    ]).transpose()

    z = np.zeros((res, res), dtype='complex128')
    
    last_cur_status_sum = res * res * 2
    has_drawn_on = 0
    for i in range(2000):
        #print('i', i)
        z = z ** 2 + c
        z[np.abs(z) > 100] = 200 # Get rid of infinity and numpy warnings
        
        cur_status = np.abs(z) < 2 #  & (img == 0)
        img[cur_status] = i
        
        cur_status_sum = np.sum(cur_status)
        #print(last_cur_status_sum - cur_status_sum)
        if last_cur_status_sum - cur_status_sum != 0:
            has_drawn_on += 1
        if (
            last_cur_status_sum - cur_status_sum < (res * res) / 10000 and
            has_drawn_on > 20
        ):
            #print(last_cur_status_sum - cur_status_sum, i)
            break
        last_cur_status_sum = cur_status_sum
        #print(np.sum(cur_status))
        #print(np.sum(cur_status))
        #if np.sum(cur_status) < res and i > 10:
        #    print('run', i, 'steps')
        #    break

    #img /= np.max(img)
    
    if np.max(img) == 0:
        print('WTF')
        return img
    
    img = img - np.min(img)
    return img / np.max(img)

pos = (0.75, 0, 3)
#img = get_img((0.75, 0.6, 1.4), 64)

# plt.imshow(img)

plt.imshow(get_img(
    (0.9064361662772606, -0.2728980630043434, 2.5 * 0.0030101738326600014),
    res=128
))
```

```python
import imageio
import matplotlib.cm

n = 0

while True:
    pos = (0.75, 0, 5)
    for i in range(30):
        pos = get_new_pos(pos, res=64)

        if i < 5:
            continue
        
        img = get_img(pos, res=8 * 4 * 256)
        
        if np.unique(img.flatten()).shape[0] < 10:
            continue
        
        cm = random.choice([
            matplotlib.cm.ocean,
            matplotlib.cm.gist_earth,
            matplotlib.cm.terrain,
            matplotlib.cm.gist_stern,
            matplotlib.cm.gnuplot,
            matplotlib.cm.gist_rainbow,
            matplotlib.cm.brg,
            matplotlib.cm.cubehelix,
            matplotlib.cm.CMRmap,
            matplotlib.cm.turbo,
            matplotlib.cm.jet,
            matplotlib.cm.gist_ncar,
            matplotlib.cm.nipy_spectral,
            matplotlib.cm.hot,
            matplotlib.cm.spring,
            matplotlib.cm.cool,
            matplotlib.cm.bwr,
            matplotlib.cm.hsv,
        ])
        
        imageio.imwrite(
            f'output/img_col_{str(n).rjust(5, "0")}.png',
            (
                (cm(1 - img) if random.random() < 0.5 else
                cm(img)
                #img
                )* 255
            ).astype('uint8')
        )
        imageio.imwrite(
            f'output/img_gray_{str(n).rjust(5, "0")}.png',
            (
                (
                img
                )* 255 * 255
            ).astype('uint16')
        )
        n += 1
        
        print('images rendered')
```
